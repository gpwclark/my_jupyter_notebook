```c
#include <string.h>
#include <stdio.h>
#include <stdlib.h>
#define SLOT_EMPTY 0
#define SLOT_FILLED 1
#define ARRAY_SIZE 256

/* C program to figure out if a given string is a set or not */

int isStringSet(char* strUnderTest)
{
    int length = strlen(strUnderTest);
    /* error checking */
    
    if (length < 2)
        return 0;
    
    char* table = (char*) malloc(sizeof(char) * ARRAY_SIZE);
    if (table == NULL) {
        fprintf(stderr, "Malloc failed.");
    }
    
    memset(table, SLOT_EMPTY, ARRAY_SIZE);
    /* error checking */
    
    int val = SLOT_EMPTY;
    for (int i = 0; i < length; i = i + 1){
        val = *(table + strUnderTest[i]);

        if (val == SLOT_FILLED)
            return 1;
        else
            *(table + strUnderTest[i]) = SLOT_FILLED;
    }
    
    free(table);
    return 0;
    
}

int main()
{

    char myString[] = "1";
    int ret = isStringSet(myString);
    
    if (ret == 1)
        printf("%s\n", "NOT a set");
    else
        printf("%s\n", "IS a set");
    
}
```
